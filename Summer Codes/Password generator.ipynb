```python
#CODE PART
#project to build a password generator 
import random
import string


#creating a password manager and password generator function
def generate_password(min_length,numbers = True,special_characters = True):
   letters = string.ascii_letters
   digits =  string.digits
   special = string.punctuation

   char = letters
   if numbers:
       char += digits

   if special_characters:
       char += special


   paswrd = ""
   meets_criteria = False
   number_present = False
   special_present = False

#creating loop for password generation    
   while not meets_criteria or len(paswrd) < min_length:
       new_char = random.choice(char)
       paswrd += new_char

       if new_char in digits:
           number_present = True

       elif new_char in special:
           special_present = True

       meets_criteria = True
       if numbers:
           meets_criteri = number_present

       if special_characters:
           meets_criteria = meets_criteria and special_present

   return paswrd
       

min_length = int(input("Enter the minimum password length:"))
number_present = input("Will password have numbers  (y/n)?").lower() == "y"
special_present = input("Will password have special characters (y/n)?").lower() == "y"
paswrd = generate_password( min_length,number_present,special_present)

print("Generated Password:",paswrd)
   
```
